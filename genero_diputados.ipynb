# Genero de los candidatos por España a las elecciones europeas de 2014

En este estudio se analiza el género del primer nombre de los candidatos de cada candidatura (se usara el término partido de forma indistinta) por España a las elecciones europeas de 2014. Se considera como "género" el género del primer nombre de cada candidato. En el estudio se usará el valor -1 para el género masculino y 1 para el femenino tal y como es ofrecido por la API de nombres consultada.

Primero se cargan los módulos necesarios para trabajar con lo datos.

```python
import pandas as pd
import numpy as np
import os
%pylab inline
```

Se leen los datos que han sido preprocesados con anterioridad tal y como se explica en el [README](https://github.com/nudomarinero/EU2014_ES_gender/) de Github.

```python
store = pd.HDFStore(os.path.join('data','candidatos.h5'))
data = store["data"]
```

## Nombres comunes

A continuación presentamos los nombres más comunes entre los candidatos.

```python
pd.value_counts(data[data["gender"] == -1.]["nombre"])[:25]
```

```python
pd.value_counts(data[data["gender"] == 1.]["nombre"])[:25]
```

## Género en los partidos

A partir de los datos se puede generar una tabla que contenga el genero de los candidatos (descartando a los suplentes). En esta tabla las filas representan a las distintas candidaturas o partidos y las columnas son los distintos candidatos en orden.

```python
genero = data[(~data["suplente"])].pivot(index='siglas', columns='numero', values='gender')
```

```python
rcParams['figure.figsize'] = (12, 8)

ny, nx = genero.shape

imshow(genero, interpolation="none")
yticks(range(ny), genero.index.values)
ylabel("Candidatura")
xticks(np.arange(nx)[::2], genero.columns.values[::2])
xlabel(u"Número de candidato")
colorbar(ticks=[-1,1], label=u"Género")
```

En la figura se muestran en dos colores (rojo - femenino; azul - masculino) el género de los candidatos de los distintos partidos. Se puede apreciar el patrón de elección de candidatos en los distintos partidos.

Podemos comprobar cuál es el género dominante dependiendo del orden del candidato. Solo se muestran los 5 primeros candidatos de cada partido para simplificar.

```python
genero.sum(axis=0)[0:5]
```

Se ve como el género de la mayoría de los cabeza de lista el género dominante es el masculino, con 27 candidatos de ventaja (33 masculino y 6 femenino). 

### Desequilibrio de género

```python
genero.sum(axis=1)
```

```python
rcParams['figure.figsize'] = (6, 4)
hist(genero.sum(axis=1), bins=np.linspace(-11,11,12))
xticks(np.linspace(-10,10,11))
xlim([-11,11])
xlabel("Desequilibrio acumulado")
ylabel("N")
```

```python
rcParams['figure.figsize'] = (6, 4)
hist(genero.ix[:,0:3].sum(axis=1), bins=[-4,-2,0,2,4])
xticks([-3,-1,1,3])
xlabel("Desequilibrio en los 3 primeros candidatos")
ylabel("N")
```

```python
rcParams['figure.figsize'] = (12, 8)
imshow(log10(genero.cumsum(axis=1).abs()), interpolation="none")
title("Desequilibrio acumulado")
yticks(range(ny), genero.index.values)
ylabel("Candidatura")
xticks(np.arange(nx)[::2], genero.columns.values[::2])
xlabel(u"Número de candidato")
from matplotlib.ticker import FuncFormatter
t = log10(np.arange(12+1))
def ffunc(a, b):
    return "%1.0f"%10**a
colorbar(ticks=t, format=FuncFormatter(ffunc), label="Cantidad de desequilibrio")
```

```python
rcParams['figure.figsize'] = (6, 4)
genero.sum(axis=0).cumsum().plot()
title("Desiquilibrio acumulado (todos los partidos)")
ylabel("Desequilibrio")
xlabel(u"Número de candidato")
```

### Desequilibrio con pesado

```python
weights_lin = ((np.arange(54, dtype=float)+1.)/54.)[::-1]
(genero*weights_lin).sum(axis=1)
```

```python
(genero*weights_lin).sum(axis=1).hist()
```

```python
weights_exp = np.exp(((np.arange(54, dtype=float)+1.)/54.)[::-1])
(genero*weights_exp).sum(axis=1)
```

```python
(genero*weights_exp).sum(axis=1).hist()
```

```python
(genero*weights_exp).sum(axis=0).plot()
```
